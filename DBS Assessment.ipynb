```python
import numpy as np
test_list=[1,3,5]
k=2

def mapping_list(test_list,k):
    test_list=sorted(test_list)
    #print(test_list)
    out=['test_list']

    for i in np.arange(1,k+1):
        if i <=1:
            pass 
        elif i ==2:
            val='test_list[{}:]'.format(1)
            out.append(val)
        else:
            val='test_list[{}:]'.format(i-1)

            out.append(val)

    re=[i.replace("'","") for i in out]

    output1=[eval(i) for i in re]
    # set returns ordered values 
    return set(zip(*output1))

result=mapping_list(test_list,k)
result
```
